```python
training=ImageDataGenerator(rescale=1/255)
vaidation=ImageDataGenerator(rescale=1/255)
```

```python
training_dataset=training.flow_from_directory('C:\\Users\\Devrim\\Desktop\\covid19_datasheet\\basedata\\training',
                                               target_size=(200,200),batch_size=3
                                               ,class_mode='binary')

vaidation_dataset=training.flow_from_directory('C:\\Users\\Devrim\\Desktop\\covid19_datasheet\\basedata\\vaidation',
                                               target_size=(200,200),batch_size=3
                                               ,class_mode='binary')
```

```python
model=tf.keras.models.Sequential([tf.keras.layers.Conv2D(16,(2,2),activation='relu',input_shape=(200,200,3)),
                            tf.keras.layers.MaxPool2D(2,2),
                            #
                            tf.keras.layers.Conv2D(32,(2,2),activation='relu'),
                            tf.keras.layers.MaxPool2D(2,2),
                            #
                            tf.keras.layers.Conv2D(64,(2,2),activation='relu'),
                            tf.keras.layers.MaxPool2D(2,2),
                            ##
                            tf.keras.layers.Flatten(),
                            ##
                            tf.keras.layers.Dense(512,activation='relu'),
                            ##
                            tf.keras.layers.Dense(1,activation='sigmoid')
    
])
```

```python

model.compile(loss='binary_crossentropy',
optimizer=RMSprop(lr=0.001),
metrics=['accuracy'])


```

```python
model_fit=model.fit(training_dataset,
                   steps_per_epoch=5,
                   epochs=30,
                   validation_data=vaidation_dataset)
```

```python
dir_path='C:\\Users\\Devrim\\Desktop\\covid19_datasheet\\basedata\\testing'
   
for i in os.listdir(dir_path):
    img=image.load_img(dir_path+'//'+i,target_size=(200,200))
    plt.imshow(img)
    plt.show()
    
    X=image.img_to_array(img)
    X=np.expand_dims(X,axis=0)
    images=np.vstack([X])
    val=model.predict(images)
    if val==0:
        print("COVID 19 HASTASI ")
    
    else:  
        print("COVID 19 HASTASI DEL")
```
